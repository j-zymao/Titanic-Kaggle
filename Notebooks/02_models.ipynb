```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('/Users/maojes/Desktop/Software Method/Titanic/Data/titanic.csv')
df
```

```python
#change male to 1 and female to 0
sex = {'male':1, 'female':0}
df.Sex = df.Sex.map(sex)
```

```python
#seperate pclass(1,2,3) in to 3 columns
pclass = pd.get_dummies(df.Pclass, prefix = 'Pclass')

df = pd.concat([df,pclass], axis = 1)
df.drop(['Pclass'], axis = 1, inplace=True)
```

```python
#categorize the titles into 
nametitle = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
```

```python
df.Name = df.Name.apply(lambda x:x.split('.')[0].strip())
df.Name = df.Name.map(nametitle)
```

```python
title = pd.get_dummies(df.Name, prefix = 'title')

df = pd.concat([df,title], axis = 1)
df.drop(['Name'], axis = 1, inplace=True)
```

```python
df
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.iloc[:,1:15],
                                                df['Survived'],test_size = 0.3, random_state = 10)
```

```python
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 10000)
```

```python
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
predict = model.predict(xtest)
predict
```

```python
accuracy_score(predict,ytest)
```

```python
pd.concat((pd.DataFrame(xtrain.columns, columns = ['element']), 
           pd.DataFrame(abs(model.coef_[0]), columns = ['relation'])), 
           axis = 1).sort_values(by='relation', ascending = False)[:15]
```
